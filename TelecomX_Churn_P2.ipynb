````python
import pandas as pd

# Expandindo a coluna 'customer' em várias colunas
customer_df = pd.json_normalize(df['customer'])

# Juntando de volta com o DataFrame original (removendo a antiga)
df = df.drop(columns=['customer']).join(customer_df)
````

````python
print(df.columns.tolist())

````

````python
dict_cols = [col for col in df.columns if isinstance(df[col].iloc[0], dict)]
print("Colunas com dict:", dict_cols)
````

````python
for col in dict_cols:
    # Normaliza o dicionário em colunas
    dict_df = pd.json_normalize(df[col]).add_prefix(f'{col}_')
    # Remove a coluna original e junta as novas colunas
    df = df.drop(columns=[col]).join(dict_df)
````

````python
for coluna in df.select_dtypes(include='object'):
    print(f"\n{coluna} - valores únicos:")
    print(df[coluna].unique()[:10])
````

````python
df['Churn'] = df['Churn'].str.lower().str.strip()
````

````python
import numpy as np

df['Churn'] = df['Churn'].replace('', np.nan)
````

````python
df = df.dropna(subset=['Churn'])
````

````python
df['Churn'] = df['Churn'].map({'yes':1, 'no':0})
````

````python
# Verificando valores únicos de todas as colunas do tipo object (categóricas)
for coluna in df.select_dtypes(include='object').columns:
    print(f"\n{coluna} - valores únicos:")
    print(df[coluna].unique())
````

````python
# Substituindo "No internet service" e "No phone service" por "No"
df.replace({
    'phone_MultipleLines': {'No phone service': 'No'},
    'internet_OnlineSecurity': {'No internet service': 'No'},
    'internet_OnlineBackup': {'No internet service': 'No'},
    'internet_DeviceProtection': {'No internet service': 'No'},
    'internet_TechSupport': {'No internet service': 'No'},
    'internet_StreamingTV': {'No internet service': 'No'},
    'internet_StreamingMovies': {'No internet service': 'No'}
}, inplace=True)
````

````python
for coluna in df.select_dtypes(include='object').columns:
    print(f"\n{coluna} - valores únicos:")
    print(df[coluna].unique())
````

````python
# Convertendo a coluna para float
df['account_Charges.Total'] = pd.to_numeric(df['account_Charges.Total'], errors='coerce')
````

````python
print(df['account_Charges.Total'].isnull().sum())
````

````python
# Remover registros com valor vazio na coluna Churn
df = df[df['Churn'] != '']
````

````python
print(df['Churn'].unique())
````

````python
# Pegando as colunas do tipo object
colunas_categoricas = df.select_dtypes(include='object').columns.tolist()

# Exibindo essas colunas
print(colunas_categoricas)
````

````python
# 1) Remover a coluna de ID, se ainda existir
if 'customerID' in df.columns:
    df = df.drop('customerID', axis=1)

# 2) Recriar a lista de colunas categóricas (tipo object), exceto o alvo
colunas_categoricas = [
    col for col in df.columns
    if df[col].dtype == 'object' and col not in ['Churn']
]

print("Colunas categóricas:", colunas_categoricas)

# 3) One‑Hot Encoding
df_encoded = pd.get_dummies(df, columns=colunas_categoricas, drop_first=True)

# 4) Separar alvo e features
y = df_encoded['Churn'].astype(int)   # garante int
X = df_encoded.drop('Churn', axis=1)

# 5) Dividir em treino e teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print("Shapes – X_train:", X_train.shape, "X_test:", X_test.shape)




````

````python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score,
                             confusion_matrix, classification_report)

# Pipeline: 1) Imputação (mediana para numéricos / mais‑frequente para binários)
#           2) Regressão Logística
log_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('model',   LogisticRegression(max_iter=1000))
])

# Treinar
log_pipeline.fit(X_train, y_train)

# Prever
y_pred_log = log_pipeline.predict(X_test)

# Avaliar
print("=== Regressão Logística (pipeline) ===")
print("Acurácia :", accuracy_score(y_test, y_pred_log))
print("Precisão :", precision_score(y_test, y_pred_log))
print("Recall   :", recall_score(y_test, y_pred_log))
print("F1‑Score :", f1_score(y_test, y_pred_log))
print("\nMatriz de Confusão:\n", confusion_matrix(y_test, y_pred_log))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred_log))

````

````python
from sklearn.ensemble import RandomForestClassifier

# Pipeline com imputação + Random Forest
rf_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('model',   RandomForestClassifier(n_estimators=100, random_state=42))
])

# Treinando
rf_pipeline.fit(X_train, y_train)

# Previsões
y_pred_rf = rf_pipeline.predict(X_test)

# Avaliação
print("=== Random Forest Classifier ===")
print("Acurácia :", accuracy_score(y_test, y_pred_rf))
print("Precisão :", precision_score(y_test, y_pred_rf))
print("Recall   :", recall_score(y_test, y_pred_rf))
print("F1‑Score :", f1_score(y_test, y_pred_rf))
print("\nMatriz de Confusão:\n", confusion_matrix(y_test, y_pred_rf))
print("\nRelatório de Classificação:\n", classification_report(y_test, y_pred_rf))
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Acessar o modelo treinado dentro do pipeline
modelo_rf = rf_pipeline.named_steps['model']

# Obter importâncias das features
importancias = modelo_rf.feature_importances_
nomes_features = X_train.columns

# Ordenar
indices_ordenados = np.argsort(importancias)[::-1]

# Visualizar as 10 mais importantes
plt.figure(figsize=(10, 6))
plt.title("Importância das Variáveis - Random Forest")
plt.barh(range(10), importancias[indices_ordenados[:10]][::-1], align='center')
plt.yticks(range(10), [nomes_features[i] for i in indices_ordenados[:10]][::-1])
plt.xlabel("Importância")
plt.tight_layout()
plt.show()
````
